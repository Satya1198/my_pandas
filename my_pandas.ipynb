## reindexing


```python
import pandas as pd

data = {
  "age": [50, 40, 30, 40],
  "qualified": [True, False, False, False]
}

idx = ["Sally", "Mary", "John", "Monica"]

df = pd.DataFrame(data, index=idx)

print("--------BEFORE--------")
print(df)

newidx = ["Robert", "Cindy", "Chloe", "Pete"]
newdf = df.reindex(newidx)

print("--------AFTER--------")
print(newdf)
```
